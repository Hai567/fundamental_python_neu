```python
total_book_price = float(input("Enter the total book price: "))
number_of_books = int(input("Enter the number of books: "))

tax = (7.5/100)*total_book_price
shipping = 2 * number_of_books

final_price = total_book_price + tax + shipping

print(f"The final price is: {final_price:.2f}")
```

```python
rented_count = input("Enter the number of movie rentals: ")
referred_count = input("Enter the number of people you referred: ")

discount = 75 if float(rented_count) + float(referred_count) > 75 else (float(rented_count) + float(referred_count))

print(f"The discount is equal to: {discount:.2f} percent")
```
